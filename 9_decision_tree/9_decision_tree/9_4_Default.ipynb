```python
import pandas as pd

default = pd.read_csv('data/Default.csv')
# 列ごとの欠損値の数
print(default.isna().sum())
# default, student のダミー変数化
df = pd.get_dummies(data=default, drop_first=True)
# データの確認
df.head(2)
```

- balance : 	クレジットカードの債務残高
- income : 	年収
- default_Yes :	クレジットカードの債務不履行
- student_Yes : 	学生

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy', max_depth=2)
Y = df.default_Yes
X = df.drop(['default_Yes'], axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_curve, auc
from sklearn.model_selection import cross_val_score
def classify(depth):
    # 決定木を作成
    model = DecisionTreeClassifier(criterion='entropy', max_depth=depth)

    # 交差検証の実行
    score = cross_val_score(model, X, Y, cv=5, scoring='roc_auc')
    print('depth={:2}: AUC={:.3f} (+/- {:.3f})'.format(depth, score.mean(), score.std()))
```

```python
classify(2)
```

```python
default_Yes = df.query('default_Yes == 1') #333 rows
default_Not_Yes333 = df.query('default_Yes == 0').head(333)
merged_df = pd.concat([default_Yes, default_Not_Yes333])
```

```python
default_Yes.info()
default_Not_Yes333.info()
merged_df.head(20)
```

```python
Y = merged_df.default_Yes
X = merged_df.drop(['default_Yes'], axis=1)

# 決定木を作成
model = DecisionTreeClassifier(criterion='entropy', max_depth=2)

# 交差検証の実行
score = cross_val_score(model, X, Y, cv=5, scoring='roc_auc')
print('depth={:2}: AUC={:.3f} (+/- {:.3f})'.format(2, score.mean(), score.std()))
```
