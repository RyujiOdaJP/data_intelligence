```python
import pandas as pd

# ItemIDã¨ã‚¢ã‚¤ãƒ†ãƒ åã®è¾æ›¸
item_dict = {
    'I1': 'ãƒ•ãƒ©ãƒ³ã‚¹ãƒ¯ã‚¤ãƒ³',
    'I2': 'ã‚¤ã‚¿ãƒªã‚¢ãƒ¯ã‚¤ãƒ³',
    'I3': 'ã‚¹ãƒšã‚¤ãƒ³ãƒ¯ã‚¤ãƒ³',
    'I4': 'ãƒãƒªãƒ¯ã‚¤ãƒ³',
    'I5': 'å›½ç”£ãƒ¯ã‚¤ãƒ³'
}

# ãƒ¯ã‚¤ãƒ³ã®è³¼è²·å±¥æ­´ã®ãƒ‡ãƒ¼ã‚¿
trans = [
    # TID: T1
    [1, 'I1'],
    [1, 'I2'],
    [1, 'I4'],
    [1, 'I5'],
    # TID: T2
    [2, 'I2'],
    [2, 'I3'],
    [2, 'I5'],
    # TID: T3
    [3, 'I1'],
    [3, 'I2'],
    [3, 'I4'],
    [3, 'I5'],
    # TID: T4
    [4, 'I1'],
    [4, 'I2'],
    [4, 'I3'],
    [4, 'I5'],
    # TID: T5
    [5, 'I1'],
    [5, 'I2'],
    [5, 'I3'],
    [5, 'I4'],
    [5, 'I5'],
    # TID: T6
    [6, 'I2'],
    [6, 'I3'],
    [6, 'I4']
]

# pandas DataFrame ã«å¤‰æ›ã€ç¢ºèª
df = pd.DataFrame(trans, columns=['TID', 'ItemID'])
df
```

- support_count(): ã‚µãƒãƒ¼ãƒˆã‚«ã‚¦ãƒ³ãƒˆã®è¨ˆç®—
- support(): ã‚µãƒãƒ¼ãƒˆã®è¨ˆç®—
- confidence(): ã‚³ãƒ³ãƒ•ã‚£ãƒ‡ãƒ³ã‚¹ã®è¨ˆç®—

```python
from functools import reduce

# TIDã‚’é‡è¤‡æ’é™¤ã®ãŸã‚é›†åˆã«å¤‰æ›
all_trans = set(df.TID)
# |D| : ãƒˆãƒ©ãƒ³ã‚¶ã‚¯ã‚·ãƒ§ãƒ³æ•°
all_trans_len = len(all_trans)

# 1ã¤ã®itemã«ã¤ã„ã¦ã€ãã®itemã‚’å«ã‚€TIDã®é›†åˆã‚’è¾æ›¸ã§ç®¡ç†
count_dict = {}
# å…¨ã¦ã®itemã«ã¤ã„ã¦ã€ãã®itemã‚’å«ã‚€TIDã®é›†åˆã‚’trans_setã«è¿½åŠ 
for item in df.ItemID.unique():
    count_dict[item] = set(df[df['ItemID'] == item].TID)

# ã‚µãƒãƒ¼ãƒˆã‚«ã‚¦ãƒ³ãƒˆã®è¨ˆç®—
# - v=1: verbose ã‚ªãƒ—ã‚·ãƒ§ãƒ³
def support_count(item_set, v=0):
    trans_set = []
    # item_setã®å„itemã«ã¤ã„ã¦ã€ãã®itemã‚’å«ã‚€TIDã®é›†åˆã‚’trans_setã«è¿½åŠ 
    for item in item_set:
        trans_set.append(count_dict[item])
    # trans_set ã®ç¢ºèªç”¨
    if (v >= 2):
        print('trans_set:', trans_set)
    # å…¨itemã‚’å«ã‚€TIDã¯ã€å„itemã‚’å«ã‚€TIDã®é›†åˆã®ç©é›†åˆ
    intersection = reduce(lambda a, x: a & x, trans_set, all_trans)
    # intersection ã®ç¢ºèªç”¨
    if (v >= 1):
        print('intersection:', intersection)
    # ã‚µãƒãƒ¼ãƒˆã‚«ã‚¦ãƒ³ãƒˆã¯ç©é›†åˆã®è¦ç´ æ•°
    return len(intersection)

# ã‚µãƒãƒ¼ãƒˆã‚«ã‚¦ãƒ³ãƒˆã®ç¢ºèª
A = {'I1'}
B = {'I3'}
support_count(A | B, v=1)
```

```python
# ã‚µãƒãƒ¼ãƒˆã®è¨ˆç®—
def support(itemset, v=0):
    return support_count(itemset, v) / all_trans_len

# ã‚³ãƒ³ãƒ•ã‚£ãƒ‡ãƒ³ã‚¹ã®è¨ˆç®—
def confidence(itemset_A, itemset_B, v=0):
    return support_count(itemset_A | itemset_B, v) / support_count(itemset_A, v)

# ã‚µãƒãƒ¼ãƒˆã€ã‚³ãƒ³ãƒ•ã‚£ãƒ‡ãƒ³ã‚¹ã®ç¢ºèª
print('{}â‡’{}: ã‚µãƒãƒ¼ãƒˆ={:.3f}, ã‚³ãƒ³ãƒ•ã‚£ãƒ‡ãƒ³ã‚¹={:.3f}'.format(
    A, B, support(A | B, v=1), confidence(A, B, v=1)))
```

## 1. Aprioriã‚¢ãƒ«ã‚´ãƒªã‚ºãƒ 

- generate_candidates(): ğ‘˜å€‹ã®è¦ç´ ã‚’æŒã¤é »å‡ºã‚¢ã‚¤ãƒ†ãƒ ã‚»ãƒƒãƒˆã®å€™è£œ ğ¶_ğ‘˜ ã‚’ä½œæˆ
- apriori(): ã™ã¹ã¦ã®é »å‡ºã‚¢ã‚¤ãƒ†ãƒ ã‚»ãƒƒãƒˆã‚’æŠ½å‡º

```python
from itertools import combinations

# Generate_Candidates(L_{k-1}, k, min_sup_count)
def generate_candidates(L, k, min_sup_count, v=0):
    Ck = []
    for l1 in L:
        for l2 in L:
            # listã®ã‚¤ãƒ³ãƒ‡ãƒƒã‚¯ã‚¹ã¯ 0 å§‹ã¾ã‚Š
            # - l1[0:n]    ã€€â€¦l1[0â€¦(n-1)] ã‚’è¿”ã™
            # - l1[0:k - 2]ã€€â€¦(k-2)ãŒ0ã®æ™‚, 0~-1ã®ç¯„å›²ã¯ãªã„ã®ã§ç©ºã®ãƒªã‚¹ãƒˆã«ãªã‚‹
            if l1[0:k - 2] == l2[0:k - 2] and l1[k - 2] < l2[k - 2]:
                # l1ã¨l2ã®å’Œé›†åˆ
                c = sorted(list(set(l1) | set(l2)))
                if (v >= 1):
                    print('k={}: {} | {} = {}'.format(k, l1, l2, c))
                # cã®k-1å€‹ã®ã‚¢ã‚¤ãƒ†ãƒ ã‹ã‚‰ãªã‚‹éƒ¨åˆ†é›†åˆã«ã¤ã„ã¦
                for i in combinations(c, k - 1):
                    # ã‚µãƒãƒ¼ãƒˆã‚«ã‚¦ãƒ³ãƒˆã‚’ç®—å‡º
                    sc = support_count(i)
                    if (v >= 2):
                        # iã¯ã‚¿ãƒ—ãƒ«ã®ãŸã‚ã€('I1',) ã®ã‚ˆã†ã«å‡ºåŠ›ã•ã‚Œã‚‹
                        print('  support_count({})={}'.format(i, sc))
                    # ã‚µãƒãƒ¼ãƒˆã‚«ã‚¦ãƒ³ãƒˆãŒ min_sup_count ã‚ˆã‚Šå°ã•ã„ã‹
                    if sc < min_sup_count:
                        # é »å‡ºã§ãªã„ãªã‚‰ã°è¿½åŠ ã—ãªã„
                        break
                else:
                    # ã•ã‚‚ãªã‘ã‚Œã°ã€Ckã«è¿½åŠ 
                    # for else æ§‹æ–‡
                    # - for ãƒ«ãƒ¼ãƒ—ã‚’ break ã§æŠœã‘ãŸæ™‚ã«ã¯ else ã¯å®Ÿè¡Œã•ã‚Œãªã„
                    # - é »å‡ºã§ãªã„ãªã‚‰ã°è¿½åŠ ã—ãªã„ãŸã‚ã€break ã™ã‚‹ã¨è¿½åŠ ã•ã‚Œãªã„
                    Ck.append(c)
    return Ck

# Apriori ã‚¢ãƒ«ã‚´ãƒªã‚ºãƒ ã®å‡¦ç†
def apriori():
    # å…¨ã¦ã®ã‚¢ã‚¤ãƒ†ãƒ 
    all_items = sorted(list(set(df.ItemID)))

    # é »å‡ºã‚¢ã‚¤ãƒ†ãƒ ã‚»ãƒƒãƒˆã®é›†åˆã‚’æ ¼ç´ã™ã‚‹dictå‹å¤‰æ•°
    L = {}

    # L1 â† 1ã¤ã®ã‚¢ã‚¤ãƒ†ãƒ ã‹ã‚‰ãªã‚‹é »å‡ºã‚¢ã‚¤ãƒ†ãƒ ã‚»ãƒƒãƒˆã®é›†åˆ
    L[1] = [[i] for i in all_items if support_count({i}) >= min_sup_count]
    # print('L1 = ', L[1])

    # k â† 2
    k = 2

    while L[k - 1] != []:
        Lk = []
        Ck = generate_candidates(L[k - 1], k, min_sup_count)
        for c in Ck:
            # print('{} {}'.format(c, support_count(c)))
            if support_count(c) >= min_sup_count:
                Lk.append(c)
        L[k] = Lk
        # print('L{} = {}'.format(k, L[k]))
        k = k + 1
    
    # å’Œé›†åˆ
    L_union = []
    for ln in L.values():
        for l in ln:
            L_union.append(l)

    return L_union
```

### Aprioriã‚¢ãƒ«ã‚´ãƒªã‚ºãƒ ã®å®Ÿè¡Œ

```python
# æœ€å°ã‚µãƒãƒ¼ãƒˆ: 0.5, æœ€å°ã‚µãƒãƒ¼ãƒˆã‚«ã‚¦ãƒ³ãƒˆ
min_sup = 0.5
min_sup_count = min_sup * all_trans_len

# Aprioriã‚¢ãƒ«ã‚´ãƒªã‚ºãƒ ã®å®Ÿè¡Œ
L_union = apriori()

# å¾—ã‚‰ã‚ŒãŸå’Œé›†åˆã®ç¢ºèª
L_union
```

## 2. ç›¸é–¢ãƒ«ãƒ¼ãƒ«ã®ç”Ÿæˆ

```python
l = ['I2', 'I3', 'I5']

# å…¨ã¦ã®çœŸéƒ¨åˆ†é›†åˆ all_s ã®ãƒªã‚¹ãƒˆã‚’åˆæœŸåŒ–
all_s = []

# å…¨ã¦ã®çœŸéƒ¨åˆ†é›†åˆ all_s ã®ä½œæˆ
for n in range(1, len(l)):
    for i in combinations(l, n):
        all_s.append(list(i))
# all_sã®ç¢ºèª
print('all_s = {}'.format(all_s))

# lã®ã‚µãƒãƒ¼ãƒˆã‚«ã‚¦ãƒ³ãƒˆ
l_support_count = support_count(l, v=0)

# å„çœŸéƒ¨åˆ†é›†åˆã«ã¤ã„ã¦ã‚³ãƒ³ãƒ•ã‚£ãƒ‡ãƒ³ã‚¹ã‚’è¨ˆç®—
for s in all_s:
    # ã‚³ãƒ³ãƒ•ã‚£ãƒ‡ãƒ³ã‚¹(s â‡’ (l-s))
    s_support_count = support_count(s, v=0)
    c = l_support_count / s_support_count
    print('{} {}/{}={}'.format(s, l_support_count, s_support_count, c))
```

```python
# æœ€å°ã‚³ãƒ³ãƒ•ã‚£ãƒ‡ãƒ³ã‚¹
min_confidence = 0.7

# ç›¸é–¢ãƒ«ãƒ¼ãƒ«ã®åˆæœŸåŒ–
assoc_rule = []

# å„çœŸéƒ¨åˆ†é›†åˆã«ã¤ã„ã¦ã‚³ãƒ³ãƒ•ã‚£ãƒ‡ãƒ³ã‚¹ã‚’è¨ˆç®—
for s in all_s:
    # ã‚³ãƒ³ãƒ•ã‚£ãƒ‡ãƒ³ã‚¹(s â‡’ (l-s))
    s_support_count = support_count(s, v=0)
    c = l_support_count / s_support_count
    print('{} {}/{}={}'.format(s, l_support_count, s_support_count, c))
    # æœ€å°ã‚³ãƒ³ãƒ•ã‚£ãƒ‡ãƒ³ã‚¹ä»¥ä¸Šã®ãƒ«ãƒ¼ãƒ«ã‚’ç›¸é–¢ãƒ«ãƒ¼ãƒ«ã«è¿½åŠ 
    if c >= min_confidence:
        #                  s â‡’ (l-s) ã‚’ãƒªã‚¹ãƒˆã§è¡¨ã™
        assoc_rule.append([s, [i for i in l if i not in s]])

# ç›¸é–¢ãƒ«ãƒ¼ãƒ«ã®ç¢ºèª
assoc_rule
```
